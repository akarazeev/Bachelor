```python
from scipy.io import arff
import pandas as pd

# filename = 'Parkinson_withoutdupl_20_v02.arff'
filename = 'HeartDisease_withoutdupl_02_v09.arff'
folder = 'HeartDisease'

data = arff.loadarff('datasets/{}/{}'.format(folder, filename))
df = pd.DataFrame(data[0])
df.columns = [x.replace(':', '_') for x in df.columns]
df.columns = [x.replace('(', '_') for x in df.columns]
df.columns = [x.replace(')', '') for x in df.columns]
df.columns = [x.replace('%', 'Percent') for x in df.columns]

df.head()
```

```python
df.to_csv('datasets/{}'.format(filename.replace('arff', 'csv')), index_label='Index')
```
